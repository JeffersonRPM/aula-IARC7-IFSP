# Random Forest 

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart_disease_health_indicators_BRFSS2015.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
df['HeartDiseaseorAttack'].unique()
```

```python
df[df['HeartDiseaseorAttack'] == 1.0]
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df_x = df.drop('HeartDiseaseorAttack', axis=1)
```

```python
X = df_x.to_numpy()
```

```python
y = df['HeartDiseaseorAttack'].to_numpy()
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3, 
                                                    random_state=42)
```

```python
y_test.shape
```

```python
X_test.shape
```

```python
classificador = RandomForestClassifier(n_estimators=100, 
                                       random_state=42)
```

```python
classificador.fit(X_train, y_train)
```

```python
y_pred_train = classificador.predict(X_train)
```

```python
accuracy_score(y_train, y_pred_train)
```

```python
y_pred_test = classificador.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_test)
```

```python
# tn, fp, fn, tp
cf_matrix = confusion_matrix(y_test, y_pred_test)
```

```python
sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')
```
