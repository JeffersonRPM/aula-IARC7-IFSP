# Jogo war simulado

```python
ataque = 0
defesa = 0 
for i in range(1000):
    jogada = np.random.choice(range(1,7), (2,3))
    jogada.sort()
    res = np.argmax(jogada, axis=0).sum()
    ataque += res
    defesa += 3 - res
print(f'Ataque: {ataque}')
print(f'Defesa: {defesa}')
```

## Teste de Hipótese

```python
a = np.array([84,72,57,46,63,76,99,91])
b = np.array([81,69,74,61,56,87,69,65,66,44,62,69])
```

```python
a.mean() 
```

```python
b.mean()
```

```python
a.mean() - b.mean()
```

```python
from statsmodels.stats.weightstats import ttest_ind

t,p, dof = ttest_ind(a ,b , alternative='larger', usevar='unequal')
print(t,p,dof)
```

```python
trials = 10000
diferencas = np.zeros(trials)
total = np.concatenate((a,b))
for i in range(trials):
    np.random.shuffle(total)
    a_new = total[:8]
    b_new = total[8:]
    diferenca = a_new.mean() - b_new.mean()
    diferencas[i] = diferenca
```

```python
px.histogram(diferencas)
```

```python
np.where(diferencas > (a.mean() - b.mean()))[0].size/trials
```

```python
np.where(diferencas > 0)[0].shape
```
